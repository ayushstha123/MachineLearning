```python
#using video
cap = cv2.VideoCapture("test.mp4")

while True:
    success, img = cap.read()
    cv2.imshow("Video", img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        cap.release()
        cv2.destroyAllWindows()
        break
```

CHAPTER 2

```python
#coverting into grayscale
imgGray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray",imgGray)
cv2.waitKey(0)
```

```python
#making the img blur
imgBlur=cv2.GaussianBlur(imgGray,(9,9),0)
cv2.imshow("blur",imgBlur)
cv2.waitKey(0)
```

```python
#edge detection(canny img)
imgCanny=cv2.Canny(img,100,100)
cv2.imshow("Canny",imgCanny)
cv2.waitKey(0)
```

```python
#making the border thicker in edges
#deine the kernel
kernel=np.ones((5,5),np.uint8)
imgDialation=cv2.dilate(imgCanny,kernel,iterations=1)
cv2.imshow("Dialation",imgDialation)
cv2.waitKey(0)
```

```python
#erossion (make the border thin)
imgErode=cv2.erode(imgDialation,kernel,iterations=1)
cv2.imshow("Erosion",imgErode)
cv2.waitKey(0)
```

CHAPTER 4 SHAPES AND TEXTS

```python
img=np.zeros((512,512,3),np.uint8)
print(img.shape)
# # img[:]=255,0,0 #to make blue
# img[200:300,100:300]=255,0,0 ##COLRING THE PORTION OF IMG
# cv2.imshow("image",img)
# cv2.waitKey(0)
```

```python
#TO CREATE LINE
cv2.line(img,(50,90),(img.shape[1],img.shape[0]),(0,255,0),3) #3 is the thickness
cv2.imshow("image",img)
cv2.waitKey(0)
```

```python
#TO CREATE AN RECTANGLE
cv2.rectangle(img,(0,0),(250,350),(0,0,255),2)
cv2.imshow("image",img)
cv2.waitKey(0)
```

```python
#TO CREATE A CIRCLE
cv2.circle(img,(400,400),30,(255,255,0),2)
cv2.imshow("image",img)
cv2.waitKey(0)
```

```python
#how to put text in img
cv2.putText(img,"OPENCV ",(200,200),cv2.FONT_HERSHEY_TRIPLEX,1,(0,150,0),2)
cv2.imshow("image",img)
cv2.waitKey(0)
```

CHAPTER 5 : WARP IMG

```python
img=cv2.imread('cards.jpg')
width,height=250,350
pts1=np.float32([[105,135],[205,60],[240,340],[320,240]])


pts2=np.float32([[0,0],[width,0],[0,height],[width,height]])
matrix=cv2.getPerspectiveTransform(pts1,pts2)
imgOutput=cv2.warpPerspective(img,matrix,(width,height))

cv2.imshow("image",img)
cv2.imshow("Output",imgOutput)
cv2.waitKey(0)
```
